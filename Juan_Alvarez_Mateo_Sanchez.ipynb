```
# Estado inicial
saldo = float(input("Ingrese saldo: "))
deposito = float(input("Ingrese deposito"))
retiro = float(input("Ingrese retiro: "))

# Actualizaci√≥n del estado
saldo = saldo + deposito   # Se deposita dinero
saldo = saldo - retiro     # Se retira dinero

# Salida del estado final
print("Saldo final de la cuenta:", saldo)
interes = 0.02 #(2% mensual)
saldo_final = saldo*interes
print(f"Saldo final con inter√©s: ${saldo_final:.2f}")
```

#### ¬øQu√© representa el estado inicial y el estado final del programa?

El estado inicial del programa representa los valores con los que comienza la ejecuci√≥n, √≥sea, el saldo de la cuenta, el monto que se va a depositar y el monto que se va a retirar (Estos datos son ingresados por el usuario). Y, el estado final muestra el resultado despu√©s de realizar la operaci√≥n  "saldo_final".


 Cada instrucci√≥n del programa modifica la memoria al crear y actualizar variables. Las instrucciones con input() guardan los valores ingresados por el usuario en memoria como lo son (saldo, deposito, retiro). Luego, las operaciones con saldo cambian su valor al sumar el dep√≥sito y restar el retiro, reemplazando el dato anterior. La variable interes almacena el porcentaje fijo, y saldo_final guarda el resultado del c√°lculo con el inter√©s.

#### EL error que nos arroja es **"TypeError: 'tuple' object does not support item assignment"** y es ddo a que la caracteristica p¬¥rincipal de unaa tupla es su inmutabilidad.

```
import math

posiciones = [(0, 0), (1, 2), (3, 4)]
total = 0

print(f"Lista de posiciones: {posiciones}\n")

for i in range(len(posiciones) - 1):

    x1, y1 = posiciones[i]
    x2, y2 = posiciones[i+1]

    x = (x2 - x1)**2
    y = (y2 - y1)**2

    distancia = math.sqrt(x + y)

    print(f"Segmento {i+1}: De {posiciones[i]} a {posiciones[i+1]} es {distancia}")

    total += distancia

print(f"\nDistancia total : {total:.4f}")
```

### $\color{green}{\text{5. Teniendo en cuenta el siguiente codigo}}$


```
from typing import Any
productos = []
while True:
    nombre = input("Ingrese producto (o 'fin' para salir): ")
    if nombre == "fin":
        break
    precio = float(input("Precio: "))
    cantidad = int(input("Cantidad: "))
    productos.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})

# Mostrar total
total = 0
for p in productos:
  if p["precio"] < 0 or p["cantidad"] < 0:
    continue
  total += p["precio"] * p["cantidad"]

print("Total del inventario:", total)


producto_mas_caro = max(productos, key=lambda p: p["precio"])
print("El producto m√°s caro es: ",producto_mas_caro)

print(productos)
```

```
import requests

# Llamado a una API p√∫blica
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
usuarios = response.json()
# Procesamiento imperativo
correos = []
nombres = []
for u in usuarios:
    nombres.append(u["name"])

    if ".biz" in u["email"]:
      correos.append((u["name"], u["email"]))

print("Usuarios encontrados:")
for n in nombres:
    print("-", n)

print("\nUsuarios cuyo correo termina en '.biz' encontrados:")
for nombre,correo in correos:
    print(f"- {nombre}: {correo}")
```

```
import requests

# Entrada de datos
nombre_pokemon = input("Ingrese el nombre de un Pok√©mon: ").lower()

# Llamado a la API
url = f"https://pokeapi.co/api/v2/pokemon/{nombre_pokemon}"
respuesta = requests.get(url)

# Verificaci√≥n del estado de la respuesta
if respuesta.status_code == 200:
    datos = respuesta.json()

    # Modelado del estado del Pok√©mon
    pokemon = {
        "nombre": datos["name"],
        "altura": datos["height"],
        "peso": datos["weight"],
        "tipos": [t["type"]["name"] for t in datos["types"]],
        "habilidades": [h["ability"]["name"] for h in datos["abilities"]]
    }

    # Salida del estado actual
    print("\nüìò Informaci√≥n del Pok√©mon")
    print(f"Nombre: {pokemon['nombre'].capitalize()}")
    print(f"Altura: {pokemon['altura']}")
    print(f"Peso: {pokemon['peso']}")
    print(f"Tipos: {', '.join(pokemon['tipos'])}")
    print(f"Habilidades: {', '.join(pokemon['habilidades'])}")

else:
    print("Pok√©mon no encontrado. Verifique el nombre ingresado.")
if len(pokemon["tipos"]) > 1:
  print("Pok√©mon de tipo mixto")

if pokemon["peso"] > 100:
  print("Pok√©mon de gran tama√±o")
else:
  print("Pok√©mon ligero")

habilidades = []
print("Habilidades:")
for skills in pokemon["habilidades"]:
  print(skills)
  if skills:
    habilidades.append(skills)

print("\nCondici√≥n adicional: ")
if len(pokemon["habilidades"]) > 1 and pokemon["peso"] < 100 and pokemon["altura"] < 100:
  print("chikito pero peligroso")

```
